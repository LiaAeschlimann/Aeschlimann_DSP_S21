### Part b - Logistic regression

```r
m1 <- glm(Direction ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume, data = df_weekly, family = binomial)
summary(m1)
contrasts(df_weekly$Direction)
```

> The predictor Lag2 (percentage return, 2 weeks previously) shows statistical significance. The estimate of Lag2 is 0.05844. It looks like when the market would return the stocks two weeks in advance, it would have a significantly better effect on the outcomes. Furthermore we have a significant Intercept value (the intercept is the mean of the response when all predictors are zero.)


### Part c - Confusion matrix

```r
glm_probs <- predict(m1, type="response")
glm_pred <- rep("Down",length(glm_probs))
glm_pred[glm_probs > 0.5] = "Up"
table(prediction =glm_pred, truth=df_weekly$Direction)
round(mean(glm_pred==df_weekly$Direction),3)
```

> From this confusion matrix we can see, that the number of correct predicions is 611 (557 + 54), which is 56.1% of correct predictions. This means, that on 54 days the stock market index would go down, as correctly predicted and on 557 days the index would go up, as correctly predicted. The number of wrong predictions is 478 (430 + 48), which results in an error rate of 43.9%. 
From this confusion matrix we can also calculate sensitivity and specificity: The true-positive rate is 92.1% (557/(557+48)) and a true-negative rate of only 11.2% (54/(54+430)) applying this model.
Furthermore, the type-I-error rate (false-positive rate or falsely predicting "Up") is 88.8% (430/(430+54)) while the type-II-error rate (false-negative rate or falsely predicting "Down") is 7.9% (54/(54+557))

### Part d - Logistic regression with Lag2 as the only predictor

```r
weekly_train <- (df_weekly$Year>=1990) & (df_weekly$Year<=2008)
weekly_test <- (df_weekly$Year>=2009)

m2 <- glm(Direction~Lag2, data=df_weekly, family=binomial, subset=weekly_train)


glm_probs2 <- predict(m2, newdata=df_weekly[weekly_test, ], type="response")
glm_pred2 <- rep("Down",length(glm_probs2))
glm_pred2[glm_probs2>.5]="Up"
table(prediction=glm_pred2, truth=df_weekly[weekly_test,]$Direction)
mean(glm_pred2==df_weekly[weekly_test,]$Direction)

```

### Part e - Repeat (d) using LDA

```r
lda1 <- lda(Direction~Lag2, data=df_weekly, subset=weekly_train)

lda_pred <- predict(lda1, newdata=df_weekly[weekly_test, ])
names(lda_pred)

table(prediction=lda_pred$class, truth=df_weekly[weekly_test,]$Direction)
mean(lda_pred$class==df_weekly[weekly_test,]$Direction)
```

### Part f - Repeat (d) using QDA

```r
qda1 <- qda(Direction~Lag2, data=df_weekly, subset=weekly_train)

qda_pred <- predict(qda1, newdata=df_weekly[weekly_test, ])
names(qda_pred)

table(prediction=qda_pred$class, truth=df_weekly[weekly_test,]$Direction)
round(mean(qda_pred$class==df_weekly[weekly_test,]$Direction),3)
```

### Part g - Repeat (d) using KNN with K = 1

```r
train = data.frame(df_weekly[weekly_train,]$Lag2) 
test = data.frame(df_weekly[!weekly_train,]$Lag2)
train_Direction = df_weekly[weekly_train,]$Direction
knn_pred <- knn(train, test, train_Direction, k=1)

table(prediction=knn_pred, truth=df_weekly[weekly_test,]$Direction)
round(mean(knn_pred==df_weekly[weekly_test,]$Direction),3)
```

### Part c - Splitting data into training and test set 

```r
length <- length(auto_new$mpg01)
random_sample1 <- sample(1:length, 3/4*length) 
random_sample2 <- (1:length)[-random_sample1]
train_auto <- auto_new[random_sample1,] 
test_auto <- auto_new[random_sample2,] 


str(train_auto)
str(test_auto)
```

### Part d - LDA on "train_auto"

> The variables that seemed most associated with mpg01 are displacement, cylinders and weight.

```r

lda_auto1 <- lda(mpg01~displacement+cylinders+weight, data=train_auto)


lda_predict <- predict(lda_auto1, newdata = test_auto)
CM2 <- table(predicted = lda_predict$class, truth = test_auto$mpg01)



length(lda_predict$class)
length(train_auto$mpg01)
mean(lda_predict$class != test_auto$mpg01)
CM2
```

> The overall test-error of this model is 9%.

### Part e - QDA on training data

```r
qda_auto1 <- qda(mpg01 ~ cylinders + displacement + weight, data = train_auto)

qda_predict <- predict(qda_auto1, newdata = test_auto)
CM3 <- table(predicted = qda_predict$class, truth = test_auto$mpg01)
CM3
```

> The overall test-error of this model is 10%.

### Part f - Logistic regression on training data

```r
lr_auto1 <- glm(mpg01 ~ cylinders + displacement + weight, data = train_auto, family = binomial)

p_auto <- predict(lr_auto1, newdata = test_auto, type = "response")
y_auto <- rep(0, length(p_auto))
y_auto[p_auto > 0.5] <- 1
CM4 <- table(predicted = as.factor(y_auto), truth = test_auto$mpg01)
CM4
```

> The overall test-error of this model is 12%

### Part g - KNN on training data

```r

train.X.Auto = data.frame(train_auto$displacement,train_auto$cylinders,train_auto$weight) 
test.X.Auto = data.frame(test_auto$displacement,test_auto$cylinders,test_auto$weight)
knn.pred.Auto <- knn(train.X.Auto, test.X.Auto, train_auto$mpg01, k=1)


table(prediction=knn.pred.Auto, truth=test_auto$mpg01)
round(mean(knn.pred.Auto==test_auto$mpg01),3)


knn.pred.Auto.k5 <- knn(train.X.Auto, test.X.Auto, train_auto$mpg01, k=5)
knn.pred.Auto.k10 <- knn(train.X.Auto, test.X.Auto, train_auto$mpg01, k=10)
knn.pred.Auto.k100 <- knn(train.X.Auto, test.X.Auto, train_auto$mpg01, k=100)
knn.pred.Auto.k200 <- knn(train.X.Auto, test.X.Auto, train_auto$mpg01, k=200)

round(mean(knn.pred.Auto.k5==train_auto$mpg01),3)
round(mean(knn.pred.Auto.k10==train_auto$mpg01),3)
round(mean(knn.pred.Auto.k100==train_auto$mpg01),3)
round(mean(knn.pred.Auto.k200==train_auto$mpg01),3)
```

> The test error of the model k=1 is 1 - 0.87 = 13%.
The test error of the model k=5 is 1 - 0.498 = 5%.
The test error of the model k=10 is 1 - 0.492 = 5.1%.
The test error of the model k=100 is 1 - 0.481 = 5.2%.
The test error of the model k=200 is 1 - 0.488 = 5.1%.
It seems like the model with *k=5* seems to fit the best, as its overall test error is the lowest.
