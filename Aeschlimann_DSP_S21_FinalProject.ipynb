```r
#Linear model with only education_parents as a predictor for each performance-test
lm0_math <- lm(formula = math_score ~ education_parents, family = binomial, data = df_performance)
print(lm0_math)
anova(lm0_math)
summary(lm0_math)

lm0_writing <- lm(formula = writing_score ~ education_parents, family = binomial, data = df_performance)
print(lm0_writing)
anova(lm0_writing)
summary(lm0_writing)

lm0_reading <- lm(formula = reading_score ~ education_parents, family = binomial, data = df_performance)
print(lm0_reading)
anova(lm0_reading)
summary(lm0_reading)
```

```r
#Linear model with education_parents and test_preperation as predictors for each performance test
lm1_math <- lm(formula = math_score ~ education_parents + test_preperation, family = binomial, data = df_performance) 
print(lm1_math)
anova(lm1_math)
summary(lm1_math)

lm1_writing <- lm(formula = writing_score ~ education_parents + test_preperation, family = binomial, data = df_performance) 
print(lm1_writing)
anova(lm1_writing)
summary(lm1_writing)

lm1_reading <- lm(formula = reading_score ~ education_parents + test_preperation, family = binomial, data = df_performance) 
print(lm1_reading)
anova(lm1_reading)
summary(lm1_reading)
```

```r
#Linear model with education_parents and test_preperation as predictors for each performance-test
lm2_math <- lm(formula = math_score ~ education_parents + test_preperation + education_parents * test_preperation, family = binomial, data = df_performance) 
print(lm2_math)
anova(lm2_math)
summary(lm2_math)

lm2_writing <- lm(formula = writing_score ~ education_parents + test_preperation + education_parents * test_preperation, family = binomial, data = df_performance) 
print(lm2_writing)
anova(lm2_writing)
summary(lm2_writing)

lm2_reading <- lm(formula = reading_score ~ education_parents + test_preperation + education_parents * test_preperation, family = binomial, data = df_performance) 
print(lm2_reading)
anova(lm2_reading)
summary(lm2_reading)
```

Befor I calculate a linear regression, I need to check the **model assumptions**:
- Linearity: The relationship between X and the mean of Y is linear
- Homoscedasticity: The variance of residual is the same for any value of X
- Independence: Observations are independent of each other
- Normality: For any fixed value of X, Y is normally distributed

```r
plot(lm1_math, which=1:4)

plot(lm1_writing, which=1:4)

plot(lm1_reading, which=1:4)
```

**Interpreting Plots** (based on, https://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/R/R5_Correlation-Regression/R5_Correlation-Regression7.html)
> Plot 1: The plot of residuals versus predicted values is useful for checking the assumption of linearity and homoscedasticity. If the model does not meet the linear model assumption, we would expect to see residuals that are very large (big positive value or big negative value). To assess the assumption of linearity we want to ensure that the residuals are not too far away from 0 (standardized values less than -2 or greater than 2 are deemed problematic). To assess if the homoscedasticity assumption is met we look to make sure that there is no pattern in the residuals and that they are equally spread around the y = 0 line.

> Plot 2: The normality assumption is evaluated based on the residuals and can be evaluated using a QQ-plot (plot 2) by comparing the residuals to "ideal" normal observations. Observations lie well along the 45-degree line in the QQ-plot, so we may assume that normality mainly hold here. Although they may be a bit off on the upper ends.

> Plot 3: The third plot is a scale-location plot (square rooted standardized residual vs. predicted value). This is useful for checking the assumption of homoscedasticity. In this particular plot we are checking to see if there is a pattern in the residuals. It looks like there are no striking patterns.

> Plot 4: The fourth plot is of "Cook's distance", which is a measure of the influence of each observation on the regression coefficients. The Cook's distance statistic is a measure, for each observation in turn, of the extent of change in model estimates when that particular observation is omitted. Any observation for which the Cook's distance is close to 1 or more, or that is substantially larger than other Cook's distances (highly influential data points), requires investigation.
In our case, the observation 60 has larger Cook's distance than other data points in Cook's distance plot, this observation also stands out in other plots in all of the three different subjects (math, reading, writing). So we may consider to remove it. It's very likely this participant was the one with 0 points in the math test.


```r
#remove observation nr. 60 out of dataframe
df_performance <- df_performance[-c(60), ]

print(df_performance)
```

> For all three perfromance tests the estimate values for the intercept is smaller, than before, with observation nr. 60 included. The intercept value is the expected mean value of the performance scores (Y) when all X=0, in our case, when education_parents and test_preperation would be 0.
 
> The slope value on the other hand represents the steepness of the function. In case of math_scores the values represent follwing: If the education_parents increases by one (one educational level higher), the math_score increases by the factor of 1.1416. And if someone does't attent the test preperation course, the math_score *decreases* in average by the factor of 5.5347.
Aswell the estimates for the slopes are less high, showing a little less strong effect of the parents educational level and a little less strong effect of absolving a test preperation course regarding math performance, than before in the model,including observation nr. 60. This makes sense, as participant nr. 60 seemed to be a very weak student, without having done any test preperation course and with their parents having "only" absolved some highschool.
